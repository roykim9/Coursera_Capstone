## Introduction
When a new business opens, it is critical to select the right neighborhood for the success. While there are many different aspects in defining a neighborhood, this project focuses on the ethnicity of the neighborhood. For example, what are the main businesses in Chinatown? How about in Little Italy? This information will be valualble to those who are starting a new business. Before deciding on where to open a business, it is important to understand where the demand is for a specific business type.

This project studied this relationship between the distribution of the ethnic restaurants and the business categories in the Bay Area, California, US.

## Methodology section

This section describes how the data is collected, processed, and presented. Each subsection is numbered with a subtitle.

### 1. Import a list of zipcodes in the Bay Area

The neighborhood and the zipcode data is imported in csv file type that was downloaded from [dataSF webpage](https://data.sfgov.org/Geographic-Locations-and-Boundaries/Bay-Area-ZIP-Codes/u5j3-svi6). The csv file is then converted into a dataframe for the ease of process in python. Unnessary columns are removed from the dataframe with only the zip code and the name of the city.

### 2. Import a list of latitude and longitude data associated with zipcodes in US

The latitude and the longitude data is downloaded from the [OpenDataSoft webpage](https://public.opendatasoft.com/explore/dataset/us-zip-code-latitude-and-longitude/export/?dataChart=eyJxdWVyaWVzIjpbeyJjb25maWciOnsiZGF0YXNldCI6InVzLXppcC1jb2RlLWxhdGl0dWRlLWFuZC1sb25naXR1ZGUiLCJvcHRpb25zIjp7fX0sImNoYXJ0cyI6W3siYWxpZ25Nb250aCI6dHJ1ZSwidHlwZSI6ImNvbHVtbiIsImZ1bmMiOiJBVkciLCJ5QXhpcyI6ImxvbmdpdHVkZSIsInNjaWVudGlmaWNEaXNwbGF5Ijp0cnVlLCJjb2xvciI6IiNGRjUxNUEifV0sInhBeGlzIjoic3RhdGUiLCJtYXhwb2ludHMiOjUwLCJzb3J0IjoiIn1dLCJ0aW1lc2NhbGUiOiIiLCJkaXNwbGF5TGVnZW5kIjp0cnVlLCJhbGlnbk1vbnRoIjp0cnVlfQ%3D%3D). Again the csv file is converted into a dataframe in python. The zip code column is renamed to 'ZIP' to match the exact column name used in the previous dataframe.

### 3. Merge the two dataframes to create a new dataframe that includes zip code, city, latitude, and longitude

Two dataframes from the previous steps are merged into one new dataframe. Again some of the columns that are not relevant in this project are removed and the final dataframe is saved with four columns: zip code, city, latitude, and longitude.

### 7. Analyze each neihborhood

The frequency of business categories are measured from each neighborhood to understand what kind of business is popular. Among all the different business categories, the restaurant categories are seperated out. ### 8. Cluster neighbors by the restaurants

In order to locate similar neighborhoods based on the restaurant presence, the k-means clustering is used. The k value is set to 10 for this project and all the neighborhoods are clustered using the restaurant data. To later evaluate the similarities between the neighborhoods, a dataframe is created with the top ten most popular restaurants in each neighborhood that includes the cluster label, latitude, and longitude data.

## Results section

Throughout the whole Bay Area, many neighborhoods have Vietnamese and Jewish restaurants as their most popular restaurants (red circle). And in the downtown areas of San Francisco and Berkeley, Vietnamese, Italian, and Japanese are popular (yellow circle). In residential areas such as Sunset, Palo Alto, Daly City, South San Francisco, and Hayward, Chinese and Mexican restaurants are shown to be popolar (skyblue circle). 


## Discussion section

Regardless of the neighborhood location, Vietnamese and Jewish restaurants has broad presence. So for a new business owner for Vietnamese and Jewish restaurants, the location in where they select is not very critical. Although further study is necessary to confirm any direct correlation between the two ethinic restuarants, the results shown here suggest there maybe some relationship between them.

It is also interesting to see the contrast between the downtown and the residential area. In the downtown of San Francisco and Berkeley, Vietnames, Italian, and Japanese are shown to be popular. It is possible that these ethinic restaurants are more popular for business meals, or a place to grab quick lunch for the workers. On the other hand, residential areas such as Sunset, Palo Alto, Daly City, South San Francisco, and Hayward, Chinese and Mexican restaurants are shown to be popular. 

For a new business, following these trends of the majority of the each ethnic restaurants would be the safe place to start. Chinese restaurants being more popular in the residential area than the downtown area is telling that there is more demand in the residential area. And if the location of the new business is fixed in one neighborhood, opening a Vietnamese or a Jewish restaurant would be a better choice as they are popular throughout the whole region. 

## Conclusion

Distribution of the ethnic restaurants in the Bay Area is studied in this project and the results show that there are specific ethnic restaurants that are popular throughout the region and there are also restaurants popular specific to the neighborhoods. This result will provide a baseline on where there are more demand for each ethnic restaurants for those who want to start a new business in restaurants.

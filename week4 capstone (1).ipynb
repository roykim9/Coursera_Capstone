# Business opportunities related to the ethnic restaurants in the Bay Area 


## Introduction
When a new business opens, it is critical to select the right neighborhood for the success. While there are many different aspects in defining a neighborhood, this project focuses on the ethnicity of the neighborhood. For example, what are the main businesses in Chinatown? How about in Little Italy? This information will be valualble to those who are starting a new business. Before deciding on where to open a business, it is important to understand where the demand is for a specific business type.

This project will study this relationship between the distribution of the ethnic restaurants and the business categories in the Bay Area, California, US.
